```python
import numpy as np          
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
np.random.seed(42)
x=(np.random.rand(100)-0.5)*100
noise=np.random.randn(100)*100

y=0.8*x**2+0.9*x+2+noise
x=x.reshape(-1,1)
y=y.reshape(-1,1)
plt.scatter(x,y, s=1, color='red', alpha=1)
plt.show()
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(x_train, y_train)
y_predict=model.predict(x_test)

plt.scatter(x,y)
sns.lineplot(x=x_test.flatten(), y=y_predict.flatten(), color='blue', label='Linear Regression')
```

```python
print(r2_score(y_test, y_predict))
```

```python
print(x_train[0])
```

```python
poly=PolynomialFeatures(degree=2)
x_train_trans=poly.fit_transform(x_train)


model=LinearRegression()
model.fit(x_train_trans, y_train)

y_predict=model.predict(poly.transform(x_test))

plt.scatter(x,y)
sns.lineplot(x=x_test.flatten(), y=y_predict.flatten(), color='blue', label='Polynomial Regression')
sns.lineplot(x=x_train.flatten(), y=model.predict(poly.transform(x_train)).flatten(), color='red', label='Polynomial Regression Train')
```

```python
print(r2_score(y_test, y_predict))
```
