```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python


x = 7 * np.random.rand(100, 1) * 10 - 2.8
y = 7 * np.random.rand(100, 1) * 10 - 2.8
z = x**3 + y**3 + 0.2*x + 0.2*y + 0.1*x*y + np.random.randn(100, 1)

data = pd.DataFrame(np.hstack((x, y, z)), columns=['x', 'y', 'z'])

X = data[['x', 'y']]
y_target = data['z']


poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y_target)



# Plotting the 3D surface
x_range = np.linspace(data['x'].min(), data['x'].max(), 50)
y_range = np.linspace(data['y'].min(), data['y'].max(), 50)
xx, yy = np.meshgrid(x_range, y_range)
grid = np.c_[xx.ravel(), yy.ravel()]
zz = model.predict(poly.transform(grid))
surface_df = pd.DataFrame({'x': grid[:, 0], 'y': grid[:, 1], 'z': zz})


fig = px.scatter_3d(data, x='x', y='y', z='z', opacity=1, color_discrete_sequence=['red'])

fig.add_traces(
    px.scatter_3d(surface_df, x='x', y='y', z='z', opacity=0.1, color_discrete_sequence=['green']).data
)

fig.update_layout(
    title="3D Polynomial Regression | Scatter + Prediction",
    scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
    margin=dict(l=0, r=0, t=30, b=0)
)
fig.show()
```

```python
print(r2_score(y_target, model.predict(X_poly)))
```
